# Data cleaning and wangling

```python
import numpy as np
import pandas as pd
import requests
```

```python
url_link ='https://finance.yahoo.com/quote/XRP-USD/history?p=XRP-USD'
r = requests.get(url_link,headers = {'User-Agent':'Mozilla/5.0(Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
read_html_pandas_data = pd.read_html(r.text)[0]
df = pd.DataFrame(read_html_pandas_data)
```

```python
df
```

```python
df.drop(100,axis = 0,inplace = True)
```

```python
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df[df['Date'] == pd.Timestamp(2022,3,25)]
```

```python
df[(df['Date'] >= pd.Timestamp(2022,3,17)) & (df['Date'] <= pd.Timestamp(2022,4,25))]
```

```python
df[(df['Date'] >= pd.Timestamp(2022,3,17)) & (df['Date'] <= pd.Timestamp(2022,4,25)) & (df['Open']<0.84)]
```

```python
col = ['Open','High', 'Low', 'Close*', 'Adj Close**', 'Volume']
for c in col:
    df[c] = pd.to_numeric(df[c])
```

```python
df.info()
```

```python
df.to_csv('XRPDATA.csv',index = False)
```

```python
df.to_excel('XRPDATA.xlsx',index = False,sheet_name = 'XRP')
```
