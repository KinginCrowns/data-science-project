```python
year = gdp['tyear']
agric = gdp['Agriculture']
ict = gdp['ICT']
educ = gdp['Education']
crude = gdp['CrudeOil']
Realestate = gdp['RealEstate']
entertainment = gdp['Art&Entertainment']
```

```python
plt.style.use('ggplot')
plt.plot(year,agric,color ='blue',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3)
plt.xlabel('YEAR')
plt.ylabel('AGRICULTRAL GDP')
plt.title('AGRICULTRAL GRP\'s IN MILLONS')
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(year,agric,color ='blue',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'AGRIC GDP')
plt.plot(year,crude,color ='red',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'CRUDE OIL GDP')
plt.plot(year,educ,color ='green',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'EDUCATION GDP')
plt.plot(year,ict,color ='orange',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'ICT GDP')
plt.xlabel('YEAR')
plt.ylabel('INDUSTRIES GDP')
plt.title('NIGERIAL GRP\'s IN MILLONS')
plt.legend()
plt.show()
```

```python
fig= plt.figure(figsize=(15,15)) # object orientend plot we specify the withd and heaih
axes = fig.add_axes([0.1,0.1,0.7,0.5]) # to alter scale of the plot
axes.plot(year,agric,color ='blue',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'AGRIC GDP')
axes.plot(year,crude,color ='red',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'CRUDE OIL GDP')
axes.plot(year,educ,color ='green',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'EDUCATION GDP')
axes.plot(year,ict,color ='orange',ls ='--',lw =3.5,marker = 'o',mfc = 'purple',mec = 'purple',markevery =3,label = 'ICT GDP')
axes.set_xlabel('YEAR',color ='green', fontsize = 30)
axes.set_ylabel('INDUSTRIES GDP',color ='green', fontsize = 30)
axes.set_title('NIGERIAL GRP\'s IN MILLONS',color ='blue', fontsize = 35)
plt.legend()
plt.show()
```

```python
fig,axes = plt.subplots(2,3,figsize =(15,8))
axes[0][0].bar(year,agric/1000000)
axes[0][0].set_xlabel('YEAR',color = 'g', fontsize = 15)
axes[0][0].set_ylabel('AGRICULTRAL GDP',color = 'g',fontsize = 15)
axes[0][0].set_title('ARICULTRAL GDP IN MILLIONS', color = 'g')
for labels in axes[0][0].get_xticklabels():
    labels.set_color('crimson')
for labels in axes[0][0].get_yticklabels():
    labels.set_color('crimson')
    
axes[0][1].bar(year,crude/1000000,color = 'gray')
axes[0][1].set_xlabel('YEAR',color = 'g', fontsize = 15)
axes[0][1].set_ylabel('CRUDE OIL GDP',color = 'g',fontsize = 15)
axes[0][1].set_title('CRUDE OIL GDP IN MILLIONS', color = 'g')
for labels in axes[0][1].get_xticklabels():
    labels.set_color('crimson')
for labels in axes[0][1].get_yticklabels():
    labels.set_color('crimson')
```
