```python
import numpy as np
import pandas as pd
```

```python
# Creating a dataframe

# using month as the column of the dataFrame

Month = ['January','Feburary','March','April','May','June','July','August','September','October','November','December']

# Days as the index of the data frame
Days = np.arange(1,31)

Readings = np.random.randint(2500,1000000,360).reshape(30,12)

Temp = pd.DataFrame(data = Readings ,index = Days, columns = Month)
Temp
```

```python
 # finding Maximun sale of March, April,September
    
    
Temp[['March','April','September']].max()
```

```python
 # finding Minimun sale of March, April,September
    
    
Temp[['March','April','September']].min()
```

```python
#  Finding the 20th and 75th percentile of all months

Temp.quantile([0.20 , 0.75])
```

```python
# Creating a new column to hold the sum of sales for each month

Temp['Total'] = Temp.sum(axis = 1)
Temp
```

```python
# Creating a new row to hold the avarage sale of each month

Temp.loc['AVG'] = Temp.mean(axis = 0)
Temp
```
