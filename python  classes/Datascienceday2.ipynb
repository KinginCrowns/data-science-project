```python
import numpy as np
import pandas as pd
```

```python
month = ['January','Feburary','March','April','May','June'] 
days = np.arange(1,31)
np.random.seed(5)
readings = np.random.randint(19,40,180)
readings = readings.reshape(30,6)


temp = pd.DataFrame(data = readings ,index = days, columns = month)
temp
```
