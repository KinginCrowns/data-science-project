```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
sns.displot(x = df['total_bill'],kde = True)
```

```python
sns.regplot(x=df['total_bill'],y = df['tip'])
```

```python
sns.regplot(x=df['total_bill'],y = df['size'])
```

```python
sns.jointplot(x=df['total_bill'],y = df['tip'],kind = 'kde')
```

```python
sns.pairplot(df)
```

```python
sns.countplot(x = df['sex'])# to check the occurence of each value and ised for cartegorical data
```

```python
sns.countplot(x = df['sex'], hue =df['time'])
plt.title('SEX AGAINST TIME')
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x = df['sex'],y = df['total_bill'],estimator = min, hue = df['time'], palette='rainbow_r')
```

```python
sns.violinplot(x = df['sex'],y = df['total_bill'], hue = df['time'], palette='rainbow_r')
```

```python
sns.swarmplot(x = df['sex'],y = df['total_bill'], hue = df['time'], palette='rainbow_r')
```

```python
sns.stripplot(x = df['sex'],y = df['total_bill'], hue = df['time'], palette='rainbow_r')
```

```python
sns.boxplot(x = df['total_bill'])
```

```python
df2 = pd.read_csv('flights.csv')
```

```python
df2.head()
```

```python
df_pv = pd.pivot_table(df2, values ='passengers', index = 'year',columns = 'month')
df_pv
```

```python
# heat maps and cluster map
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(df_pv, annot = True , fmt = '.5g')
```

```python
plt.figure(figsize = (15,10))
sns.clustermap(df_pv, annot = True , fmt = '.5g')
```
